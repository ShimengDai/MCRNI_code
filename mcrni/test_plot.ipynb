```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Show the plot
plt.show()
```

```python

# First line
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]

# Second line
y2 = [1, 3, 5, 7, 9]

# Plot both lines
plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")

# Labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Show legend
plt.legend()

# Show plot
plt.show()
```

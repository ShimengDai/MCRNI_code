# MCRNI Single Model 

```python
from scipy.stats import rankdata 
import numpy as np
from MCRNI import MCRNI
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
```

```python
#calculate rank of each item in array
my_array = np.array([3, 5, 2, 1, 1, 9, 9, 9])
ranks = rankdata(my_array, method='ordinal') 
print("Array:", my_array)
print("Rank:", ranks)
```

# Full Model 

```python
# Logistic Regression

df = pd.read_csv("/Users/shimengdai/Dropbox/Michigan State/Papers/MCRNI/mcrni/input_data/predictions_wide_full.csv")
y_true = df["dropout"] 

print("Full Model:", "Logistic Regression")
y_scores = df["pred_logreg_proba"] 
model = MCRNI(y_true, y_scores)
model.__rank__()  # Compute rank sums and U stats
model.compute_mcrni_with_auc(0.5)  # Compute MCRNI using AUC=0.5
scrni = model.scrni()
model.print_report(threshold=0.2)
```

```python

print("Full Model:", "Random Forest")
y_scores = df["pred_rf_proba"] 
model = MCRNI(y_true, y_scores)
model.__rank__()  # Compute rank sums and U stats
model.compute_mcrni_with_auc(0.5)  # Compute MCRNI using AUC=0.5
scrni = model.scrni()
model.print_report(threshold=0.2)
```

```python
print("Full Model:", "SVM")
y_scores = df["pred_svm_proba"] 
model = MCRNI(y_true, y_scores)
model.__rank__()  # Compute rank sums and U stats
model.compute_mcrni_with_auc(0.5)  # Compute MCRNI using AUC=0.5
scrni = model.scrni()
model.print_report(threshold=0.2)
```
